```python
import geopandas as gpd
import pandas as pd
import json
import plotly.express as px
import plotly.graph_objs as go

geodf = gpd.read_file('C:\\Users\\mikha\\Box\\Nile_basin\\Nile_basin_shapefiles\\eLVB_61subBasins.shp')
```

```python
# Converts a shapefile into a json file, useable by Plotly/Mapbox

geodf.to_file("path_to_GeoJSON _file", driver = "GeoJSON")
with open("path_to_GeoJSON _file") as geofile:
    j_file = json.load(geofile)

i=1
for feature in j_file["features"]:
    feature ['id'] = str(i).zfill(2)
    i += 1
```

```python
L = len(j_file['features'])
```

```python
j_file['features'][0].keys()
```

```python
j_file['features'][7]['properties']
```

```python
#'id' is a two-digit string, for example, '01'

locations = ["{0:0=2d}".format(i+1) for i in range(L)]
text = [feat['properties']['Id']  for feat in j_file['features'] if feat['id'] in locations] #province names  
```

```python
z = [i for i in range(L)]
```

```python
 trace = go.Choroplethmapbox(z=z,
                            locations=locations,
                            colorbar=dict(thickness=20, ticklen=3),
                            geojson=j_file,
                            text=text,
                            hoverinfo='all',
                            marker_line_width=0.1, marker_opacity=0.7)
                            
                            
layout = go.Layout(title_text= 'Choroplethmapbox',
                   title_x=0.5, width = 700, height=700,
                   mapbox = dict(
                                 accesstoken= 'pk.eyJ1Ijoic21pbG92aWMiLCJhIjoiY2s5Y3Z0YjVtMDJyZzNtcGpycHpuYnJqZyJ9.6lSmRCspWDfcvSXaggZdWQ',
                                 zoom=2.650,
                               ))

fig = go.Figure(data=[trace], layout =layout)
fig.update_layout(mapbox_style = "basic")
fig.show()
```
